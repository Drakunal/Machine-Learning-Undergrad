#### Feature Selection

```python
x = data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = data['Species']
```

```python
x
```

```python
y
```

#### Splitting of Dataset

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(x,y)
```

```python
train_x
```

```python
train_y
```

```python
test_x
```

#### Training of the machine learning model

```python
from sklearn.svm import SVC

model = SVC()
model.fit(train_x,train_y)
```

#### Prediction

```python
model.predict([[5.1,3.5,1.4,0.2]])
```

```python
prediction = model.predict(test_x)
```

```python
prediction
```

#### Evaluation

```python
from sklearn.metrics import accuracy_score

accuracy_score(test_y, prediction)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(test_y, prediction)
```

```python


cm = confusion_matrix(test_y, prediction)

# Transform to df for easier plotting
cm_df = pd.DataFrame(cm,
                     index = ['setosa','versicolor','virginica'],
                     columns = ['setosa','versicolor','virginica'])

plt.figure(figsize=(5.5,4))
sns.heatmap(cm_df, annot=True)
plt.title('SVM Linear Kernel \nAccuracy:{0:.3f}'.format(accuracy_score(test_y, prediction)))
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(test_y, prediction, average='macro')
recall = recall_score(test_y, prediction, average='macro')
f1 = f1_score(test_y, prediction, average='macro')

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```
